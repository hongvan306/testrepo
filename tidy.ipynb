```python
ebola_long['variable'].str.split('_', expand=True)
```

```python
# create new columns
# khi mà một câu lệnh quá dài, có thể dùng (), hoặc xuống dòng trước các dấu "."
ebola_long [['stats_e', 'country_e']] = (ebola_long['variable']
                                         .str
                                         .split('_', expand=True))
```

```python
# to reduce the duplicates, the size of storage
billboard_songs = billboard_songs.drop_duplicates()
```

```python
billboard_songs.shape
```

```python
# give a number, generate a value
# tạo ra cột id
billboard_songs['id'] = range(len(billboard_songs))
```

```python
billboard_songs.head()
```

```python
# nếu ko ghi index= False thì cái cột đầu tiên sẽ bị save luôn 
# save cái billboard_songs thành 1 file csv 
billboard_songs.to_csv('billboard_songs.cvs', index= False )
```

```python
billboard_ratings = billboard_melt.merge(
    billboard_songs, on=['year','artist', 'track','time']
)
```

```python
# chỉ quan tâm về một số cột cần thiết thôi
billboard_ratings = billboard_ratings [['id','date.entered','week','rating']]
```
